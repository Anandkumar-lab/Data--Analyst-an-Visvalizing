```python
# the data show review per month
plt.plot(Air['reviews_per_month'],color ='g',linewidth=2)
plt.title('Reviews per month')
plt.show()
```

```python
# the data last review column show
sns.histplot(Air['last_review'],bins =20)
plt.title('last Review')
plt.show()
```

```python
# the data show id and Review per months or Room type columns 
sns.lineplot(x =Air['id'],y =Air['reviews_per_month'],hue=Air['room_type'], data =Air)
plt.title('Room type')
plt.show()
```

```python
# the data show column Host id and Host name
sns.pointplot(x=Air['host_id'],y =Air['host_name'],data=Air)
plt.title('Host Id and Host Name')
plt.show()
```

```python
# the data print column name room type,price,availability show
grid1 = sns.FacetGrid(Air,col='room_type')
grid1.map(plt.hist,'price')
grid3=sns.FacetGrid(Air,col='room_type')
grid3.map(plt.scatter,'price','availability_365')
plt.show()
```

```python
# the data print Minimum night column
sns.violinplot(x=Air['room_type'],y=Air['minimum_nights'],data=Air)
plt.title(' Night Room')
plt.show()
```

```python
# the data print id,neighbourhood column show
fig=px.bar(Air,x='number_of_reviews',y='id',color='neighbourhood_group',)
fig.show()
```

```python
# the data print neighbourhood group
fig=px.box(Air,x='room_type',y='longitude',color='neighbourhood_group',notched=True)
fig.show()
```

```python
# the data print room type show 
fig=px.strip(Air,x='price',y='availability_365',orientation='v',color='room_type')
fig.show()
```
